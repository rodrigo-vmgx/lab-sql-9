```python
#1
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
result = engine.execute("SELECT COUNT(DISTINCT SUBSTRING_INDEX(last_name, ' ', -1)) FROM actor")
pd.DataFrame(result)
```

```python
#2
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
result = engine.execute("SELECT *, CASE WHEN DAYOFWEEK(rental_date) IN (1,7) THEN 'weekend' ELSE 'workday' END AS day_type FROM rental")
pd.DataFrame(result)
```
